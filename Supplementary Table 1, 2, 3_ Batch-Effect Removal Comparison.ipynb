```python
import pandas as pd
import numpy as np
from scipy.stats import sem, t

# Confidence level
confidence = 0.95

def mean_margin(series, confidence=0.95):
    n = len(series)
    mean = np.mean(series)
    se = sem(series)
    margin = se * t.ppf((1 + confidence) / 2., n - 1) if n > 1 else 0.0
    return f"{mean:.4f} Â± {margin:.4f}"

def summarize_auc(df, method_name):
    grouped = df.groupby("PROBLEM_KEY")
    auc_summary = grouped["auc"].agg(lambda x: mean_margin(x, confidence))
    auc_summary.name = method_name
    return auc_summary

method_tables = {
    "Harmony": scpss_metrics,
    "Scanorama": scanorama_metrics,
    "scVI": scvi_metrics,
    "No Integration": pca_metrics
}

summarized = [summarize_auc(df, method) for method, df in method_tables.items()]
final_df = pd.concat(summarized, axis=1).reset_index()

ordered_keys = [
    "ref_0_que_1hr",
    "ref_0_que_4hr",
    "ref_0_que_1",
    "ref_0_que_3",
    "ref_0_que_7",
    "ref_0_que_1-7"
]

final_df["PROBLEM_KEY"] = pd.Categorical(final_df["PROBLEM_KEY"], categories=ordered_keys, ordered=True)
final_df = final_df.sort_values("PROBLEM_KEY").reset_index(drop=True)
final_df.to_csv(f"{OUT_PATH}/auc_integration.csv", index=False)
final_df
```
