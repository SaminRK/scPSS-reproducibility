```python
import pandas as pd
import numpy as np
from scipy.stats import sem, t


# Confidence level
confidence = 0.95

# Function to calculate mean ± margin
def mean_margin(series, confidence=0.95):
    n = len(series)
    mean = np.mean(series)
    se = sem(series)
    margin = se * t.ppf((1 + confidence) / 2., n - 1)
    return f"{mean:.4f} ± {margin:.4f}"

def aggregate(metrics):

    # Group by 'Group' and calculate mean ± margin for other columns
    grouped = metrics.groupby('PROBLEM_KEY')

    result = grouped.agg({
        col: lambda x: mean_margin(x, confidence)
        for col in metrics.columns if col != 'PROBLEM_KEY'
    })

    # Reset index to make it look cleaner
    result.reset_index(inplace=True)

    return result
```

```python
import pandas as pd
from functools import reduce

# List of dataframes and corresponding prefixes
dfs = [df_contrastivevi_results_var1, df_contrastivevi_results_var2, df_hidden_results, df_multimil_results, df_scpss_results, ]
prefixes = ['contrastivevi_var1', 'contrastivevi_var2', 'hidden', 'multimil',  'scpss', ]
```

```python
# Step 1: Select PROBLEM_KEY and aupc, rename aupc with prefix
renamed_dfs = [
    df[['PROBLEM_KEY', 'auc']].rename(columns={'auc': f'{prefix}_auc'})
    for df, prefix in zip(dfs, prefixes)
]

# Step 2: Merge all on PROBLEM_KEY
merged_df = reduce(lambda left, right: pd.merge(left, right, on='PROBLEM_KEY', how='outer'), renamed_dfs)

# Step 3: Set PROBLEM_KEY as index
merged_df.set_index('PROBLEM_KEY', inplace=True)

custom_order = [
    'ref_0_que_1hr',
    'ref_0_que_4hr',
    'ref_0_que_1',
    'ref_0_que_3',
    'ref_0_que_7',
    'ref_0_que_1-7',
]

merged_df = merged_df.reindex(custom_order)
merged_df.to_csv("/content/drive/MyDrive/Research - Sohel sir/Pathological Cells/Pathological Cell Identification/reproducibility/outputs/Table1.csv")
merged_df
```

```python
import pandas as pd

# Assuming df is your existing DataFrame (with ± in values)
# Step 1: Extract the mean part before '±' and convert to float
df_numeric = merged_df.map(lambda x: float(str(x).split('±')[0].strip()))

# Step 2: Get scpss_auc separately
scpss = df_numeric["scpss_auc"]

# Step 3: Calculate % improvement over each method (excluding scpss itself)
baseline_methods = df_numeric.drop(columns="scpss_auc")
improvement = (scpss.values.reshape(-1, 1) - baseline_methods.values) / baseline_methods.values * 100

# Step 4: Create DataFrame with improvements
improvement_df = pd.DataFrame(improvement, columns=baseline_methods.columns, index=merged_df.index)

all_improvements = improvement_df.values.flatten()

# Step 5: Compute overall median % improvement
overall_median_improvement = pd.Series(all_improvements).median()

# Result
print(f"Overall median % improvement of scPSS over all other methods: {overall_median_improvement:.2f}%")
```
